```python
def tent_function(a, b, x=4, y=6, delta=2):
    tent = 1 - (1/delta)*max(abs(x-a), abs(y-b))
    tent = max(0,tent)
    return(tent)
```

```python
### Create linspace and meshgrid for plotting
import numpy as np
n = 100
x = np.linspace(0, 10, n)
y = np.linspace(0, 10, n)
xv, yv = np.meshgrid(x, y)
```

```python
### Density of Tent Function
tent_output = np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        tent_output[i,j]=tent_function(xv[i,j], yv[i,j])
        
plt.pcolormesh(x, y, tent_output)
plt.grid(visible=True, c='white')
```

```python
n = 100
x = np.linspace(0, 10, n)
y = np.linspace(0, 20, n)
xv, yv = np.meshgrid(x, y)

tent_output = np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        tent_output[i,j]=tent_function(xv[i,j], yv[i,j]-xv[i,j])
        
plt.pcolormesh(x, y, tent_output)
plt.xlim(0,10)
plt.ylim(0,20)
plt.xticks(np.arange(0, 11, step=2))
plt.yticks(np.arange(0, 21, step=2))

x_bins = np.linspace(0, 10, 11)
y_bins = np.linspace(0, 10, 11)

t = np.linspace(0,10, 100)

for dis in y_bins:
    plt.plot(t, t+dis, c="white")

for dis in x_bins:
    plt.plot(dis*np.ones(len(t)), t+dis, c="white")
```
